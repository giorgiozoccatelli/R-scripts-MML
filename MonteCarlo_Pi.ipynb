```r
### a function to estimate pi based on nsim i.i.d. runs
estimatepi <- function(nsim){
# x is a matrix of nsim rows and two columns, all uniform [-1,1]
x <- matrix(runif(2*nsim,-1,1), nrow=nsim, ncol=2, byrow=T)
# check row-by-row if inside the circle
checkcircle <- function(z) (z[1]^2+z[2]^2<=1)
x <- cbind(x, apply(x, 1, checkcircle))
# returning how many in the circle
4*sum(x[,3])/nsim
}
### estimate pi based on 10Ë†5 simulations
estimatepi(10^5)
```
